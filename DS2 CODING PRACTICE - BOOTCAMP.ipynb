```python
#return all beer servings less than 89 AND (&) from the EU
less_than_89 = (drinks.beer_servings < 89) & (drinks.continent =='EU')
drinks[less_than_89].head()
```

```python
# Use the pipe | as an OR operator
less_than_89 = (drinks.beer_servings < 89) | (drinks.continent =='EU')
drinks[less_than_89].head()
```

```python
less_than_89.index
```

```python
# the OR operator takes priority over the &
less_than_100 = (drinks.beer_servings <=100) & (drinks.continent == 'AF') | (drinks.wine_servings > 200)
drinks[less_than_100].head()
```

```python
#the long way of selecting occupations
multiple_occupations = (users.occupation == 'student') | (users.occupation =='doctor') | (users.occupation =='lawyer')
users[multiple_occupations].head()
```

```python
# the quick way of returning multiple conditions - isin
lawstudoc = users[users.occupation.isin(['student', 'doctor','lawyer'])]
lawstudoc.head()
```
